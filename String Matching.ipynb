```python
def func(value):
    return ''.join(value.splitlines())
```

```python
#Read All Data
docname = []
doctext = []
for i in range(1, 11):
    f = open("Research#" + str(i) + '.txt', 'r', encoding='UTF-8')
    docname.append("Research#" + str(i) + '.txt')
    doctext.append(func(f.read()))
```

```python
def Brute_Force(text, pattern):
    if (len(text)<len(pattern)):
        return False
    
    for i in range(len(text)):
        for j in range(len(pattern)):
            if i + j >= len(text):
                break
            if text[i + j] != pattern[j]:
                break
        else:
            return True
    return False
```

```python
def Rabin_Karp(text, pattern):
    if (len(text)<len(pattern)):
        return False
    d = 10
    q = 13
    m = len(pattern)
    n = len(text)
    p = 0
    t = 0
    h = 1
    i = 0
    j = 0

    for i in range(m-1):
        h = (h*d) % q

    # Calculate hash value for pattern and text
    for i in range(m):
        p = (d*p + ord(pattern[i])) % q
        t = (d*t + ord(text[i])) % q

    # Find the match
    for i in range(n-m+1):
        if p == t:
            for j in range(m):
                if text[i+j] != pattern[j]:
                    break

            j += 1
            if j == m:
                return True

        if i < n-m:
            t = (d*(t-ord(text[i])*h) + ord(text[i+m])) % q

            if t < 0:
                t = t+q
    return False
```

```python
def KMP( text,pattern):
    if (len(text)<len(pattern)):
        return False
    a = len(text)
    b = len(pattern)
    prefix_arr = get_prefix_arr(pattern, b)
  
    initial_point = []
    m = 0
    n = 0
  
    while m != a:
       
        if text[m] == pattern[n]:
            m += 1
            n += 1
      
        else:
            n = prefix_arr[n-1]
       
        if n == b:
            initial_point.append(m-n)
            n = prefix_arr[n-1]
        elif n == 0:
            m += 1
   
    return (True if len(initial_point) > 0 else False)
def get_prefix_arr(pattern, b):
    prefix_arr = [0] * b
    n = 0
    m = 1
    while m != b:
        if pattern[m] == pattern[n]:
            n += 1
            prefix_arr[m] = n
            m += 1
        elif n != 0:
                n = prefix_arr[n-1]
        else:
            prefix_arr[m] = 0
            m += 1
    return prefix_arr
```

```python
def Search_Function(searchword="virus", algorithm="Brute Force", MatchWord = False, MatchCase = False):
    
    if (algorithm == "BF"):
        algorithm = Brute_Force
    if (algorithm == "KMP"):
        algorithm = KMP
    if (algorithm == "RK"):
        algorithm = Rabin_Karp
        
    docslist = []
    if (MatchWord == False and MatchCase == False):
        lowered_text = [x.lower() for x in doctext]
        searchword = searchword.lower()
        for i in range(0, len(docname)):
            if (algorithm(lowered_text[i], searchword)):
                docslist.append(docname[i])
    
    if (MatchWord == False and MatchCase == True):
        for i in range(0, len(docname)):
            if (algorithm(doctext[i], searchword)):
                docslist.append(docname[i])
                
    elif (MatchWord == True and MatchCase == False):
        lowered_text = [x.lower() for x in doctext]
        searchword = searchword.lower()
        for i in range(0, len(docname)):
            wordslist = lowered_text[i].split(' ')
            for j in range(0, len(wordslist)):
                if (algorithm(wordslist[j], searchword)):
                    docslist.append(docname[i])
                    break
    
    elif (MatchWord == True and MatchCase == True):
        for i in range(0, len(docname)):
            wordslist = doctext[i].split(' ')
            for j in range(0, len(wordslist)):
                if (algorithm(wordslist[j], searchword)):
                    docslist.append(docname[i])
                    break
                    
    return docslist
```
